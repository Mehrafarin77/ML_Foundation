The Problem: A robber is escaping from a sheriff. The robber has got speed of 150 km/h as the sheriff has got speed of 180 km/h but 5 minutes late.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000)      # start, end, points
```

I convert speeds to km/min because of the 5-minutes of being late.

```python
robber_distance = 2.5 * t
sheriff_distance = 3 * (t - 5)
```

```python
fig, ax = plt.subplots()
plt.title('A Bank Robber Caught')
plt.xlabel('Time (in minutes)')
plt.ylabel('Distance (in Km)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, robber_distance, c='green')
ax.plot(t, sheriff_distance, c='brown')
plt.axvline(x= 30, color='purple', linestyle='--')
_ = plt.axhline(y= 75, color='purple', linestyle='--')

plt.show()
```
